```python
import librosa
import librosa.display
import soundfile as sf
from soundfile import SoundFile
import matplotlib.pyplot as plt
import numpy as np

filename = '/Users/andresharisaduvvuri/Documents/GitHub/groupMLProject/ML Brostep/Purge.wav'

y, sr = librosa.load(filename, sr = 44000)
librosa.display.waveplot(y, sr=sr)
```

```python
tempo, beat_frames = librosa.beat.beat_track(y=y, sr=sr)
print('Estimated tempo: {:.2f} beats per minute'.format(tempo))
```

```python
beat_times = librosa.frames_to_time(beat_frames, sr = sr)
```

```python
S = librosa.feature.melspectrogram(y=y,sr=sr, n_mels=128,fmax=15000)
```

```python
S_dB = librosa.power_to_db(S, ref=np.max)
```

```python
fig, ax = plt.subplots()
img = librosa.display.specshow(S_dB, x_axis='time',y_axis='mel', sr=sr, fmax=15000,ax=ax)
fig.colorbar(img,ax=ax, format='%+2.0f dB')






```

```python
yy = librosa.feature.inverse.mel_to_audio(S, sr=sr)


#sf.write('/Users/andresharisaduvvuri/Documents/GitHub/groupMLProject/ML Brostep/test/Purgee.wav',yy,samplerate=sr)

```
